Processing, cleaning, etc

```python
# Change column type to datetime64[ns] for column: 'date'
df = df.astype({'date': 'datetime64[ns]'})

# Changing the type from objetc to numeric for columns that are numeric
df['male_smokers'] = pd.to_numeric(df['male_smokers'], errors='coerce')
df['female_smokers'] = pd.to_numeric(df['female_smokers'], errors='coerce')


# Drop missing values from continent column

df = df.dropna(subset=['continent'])

# Drop columns that will not be used in the analysis
df = df.drop(columns=['total_cases_per_million', 'new_cases_per_million', 'total_deaths_per_million', 'new_deaths_per_million', 
                      'total_tests', 'new_tests', 'positive_rate', 'tests_units', 'total_vaccinations_per_hundred', 'people_vaccinated_per_hundred', 
                      'people_fully_vaccinated_per_hundred', 'gdp_per_capita','aged_70_older' ])



```

```python
# Function to remove the last zero from a float number and handle NaN in specifics columns
# This is necessary due to an encoding issue in the table that affected some columns

def remove_ultimo_zero(num):
    if isinstance(num, (int, float)) and not numpy.isnan(num) and numpy.isfinite(num):
        return int(num / 10)
    else:
        return 0  # Replace with another value if desired

# Applying the function to the columns
df['total_cases'] = df['total_cases'].apply(remove_ultimo_zero)
df['population'] = df['population'].apply(remove_ultimo_zero)
df['new_cases'] = df['new_cases'].apply(remove_ultimo_zero)
df['total_deaths'] = df['total_deaths'].apply(remove_ultimo_zero)
df['new_deaths'] = df['new_deaths'].apply(remove_ultimo_zero)

# Converting columns to integers
df['total_cases'] = df['total_cases'].astype(int)
df['population'] = df['population'].astype(int)
df['new_cases'] = df['new_cases'].astype(int)
df['total_deaths'] = df['total_deaths'].astype(int)
df['new_deaths'] = df['new_deaths'].astype(int)

# Convert columns to numeric, handling errors by coercing them to NaN
# Fill missing (NaN) values with 0
# Convert to integer type
df['aged_65_older'] = pd.to_numeric(df['aged_65_older'], errors='coerce').fillna(0).astype(int)
df['female_smokers'] = pd.to_numeric(df['female_smokers'], errors='coerce').fillna(0).astype(int)
df['male_smokers'] = pd.to_numeric(df['male_smokers'], errors='coerce').fillna(0).astype(int)
df['female_smokers'] = pd.to_numeric(df['female_smokers'], errors='coerce').fillna(0).astype(int)
df['life_expectancy'] = pd.to_numeric(df['life_expectancy'], errors='coerce').fillna(0).astype(int)
df['cardiovasc_death_rate'] = pd.to_numeric(df['cardiovasc_death_rate'], errors='coerce').fillna(0).astype(int)
df['diabetes_prevalence'] = pd.to_numeric(df['diabetes_prevalence'], errors='coerce').fillna(0).astype(int)



```

SCHEMA

```python
# Validating schema with pandera
# Create a dictionary to store column schemas
schema_dict = {}

# Iterate through all columns in the DataFrame
for column, dtype in df.dtypes.items():
    if dtype == 'object':
        # If the data type of the column is 'object', validate as STRING and allow null values
        schema_dict[column] = pa.Column(pa.String, nullable=True)
    elif dtype == 'int64':
        # If the data type of the column is 'int64', validate as INT and allow null values
        schema_dict[column] = pa.Column(pa.Int, nullable=True)
    elif dtype == 'float64':
        # If the data type of the column is 'float64', validate as FLOAT and allow null values
        schema_dict[column] = pa.Column(pa.Float, nullable=True)
    elif dtype == 'datetime64[ns]':
        # If the data type of the column is 'datetime64[ns]', validate as DATETIME and allow null values
        schema_dict[column] = pa.Column(pa.DateTime, nullable=True)

# Create a Pandera schema based on the dictionary
schema = pa.DataFrameSchema(schema_dict)

# Validate the DataFrame based on the schema
try:
    schema.validate(df)
    print("The DataFrame is valid with the new schema.")
except pa.errors.SchemaError as e:
    print("The DataFrame is not valid. Validation errors:")
    print(e)
```

GRAPH

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Assuming 'df' is your DataFrame with the data
# Make sure the data is loaded into the variable 'df'

# Select only numeric columns
numeric_data = df.select_dtypes(include=['float64', 'int64'])

# Calculate the correlation matrix
correlation_matrix = numeric_data.corr()

# Filter for strong correlations (>= 0.5 or <= -0.5)
strong_correlations = correlation_matrix[(correlation_matrix >= 0.5) | (correlation_matrix <= -0.5)]

# Create the heatmap of correlations
plt.figure(figsize=(10, 8))
sns.heatmap(strong_correlations, annot=True, cmap='coolwarm', fmt=".2f")

# Add title
plt.title("Correlation Heatmap (|Correlation| >= 0.5)")

# Show the plot
plt.show()
```
